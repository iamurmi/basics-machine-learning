```python
#Matplotlib is a python library used for data visualization.
#You can create bar-plots, scatter-plots, histograms and a lot more with matplotlib.
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
#Line plot
y= 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y = 5*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Adding Title and labels
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x.label")
plt.ylabel("y.label")
plt.show()
```

```python
#Changing Line Aesthetics
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('x vs y')
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.show()
```

```python
#Adding two lines in the same plot
x=np.arange(1,11)
y1=2*x
y2=5*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':')
plt.plot(x,y2,color='r',linewidth=3)
plt.title("Two lines in a plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
#Adding sub-plots
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':')
plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linewidth=3)
plt.title("Two lines in a plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
#Bar Plot
student = {"Urmi":87,"Mira":56,"Sita":97,"Tara":47}
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
#Adding Title and labels
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Horizontal Bar Plot
plt.barh(names,values,color='g')
plt.title('Distribution of student marks')
plt.xlabel('Names of Students')
plt.ylabel('Marks of Students')
plt.grid(True)
plt.show()
```

```python
#Scatter Plot
#Creating a basic scattr-plot
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
#Changing Mark Aesthetics
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker="*",c="r",s=100)
plt.show()
```

```python
#Adding two markers in the same plot
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker="*",c="r",s=100)
plt.scatter(x,b,marker=".",c="g",s=200)
plt.show()
```
